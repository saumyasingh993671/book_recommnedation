## Popularity Based Recommender system

```python
rating_with_name=rating.merge(book,on='ISBN')
```

```python
num_rating_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

```python
popular_df=popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
final_popular_df=popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
final_popular_df
```

## Collaborative Filtering Based Recommender System

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users=x[x].index
```

```python
filter_rating=rating_with_name[rating_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y=filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filter_rating[filter_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
similarity_score.shape
```

```python
index=np.where(pt.index=='The Fellowship of the Ring ( The lord of the Rings, Part 1)')[0][0]
```

```python
def recommend(book_name):
    #index fetch
    index=np.where(pt.index==book_name)[0][0]
#distances=similarity_score[index]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df=book[book['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M']))
        
        data.append(item)
    return data
    
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Notebook')
```

```python
recommend('The Da Vinci Code')
```

```python
recommend('Harry Potter and the chapter of secrets (Book 2)')
```
